## Cartopy Explained

### What is it?
#### A python package which provides a set of tools for creating projection-aware geospatial plots using python's standard plotting package (__matplotlib__). Cartopy also has a robust set of tools for defining projections and reprojecting data :)

### Building Blocks of Cartopy

### 1. Projections (cartopy.crs)
#### You are able to define and transform data to fit a certain cartographic projection.
##### Mercator, Lambert Conformal, and more!
### 2. Features (cartopy.feature)
#### A convient set of data loaders for adding context to maps.
##### Coastlines, borders, and. more!

```python
#Example from scitools.org showing a simple map that compares geodetic and Plate Carree lines between two locations
def main():
    fig = plt.figure(figsize=(10, 5))
    ax = fig.add_subplot(1, 1, 1, projection=ccrs.Robinson())

    # make the map global rather than have it zoom in to
    # the extents of any plotted data
    ax.set_global() #shows globe

    ax.stock_img() #shows globe with colors
    ax.coastlines() #adds in coastlines to globe

    ax.plot(-0.08, 51.53, 'o', transform=ccrs.PlateCarree()) #location of plotting a line based on lat and long
    ax.plot([-0.08, 132], [51.53, 43.17], transform=ccrs.PlateCarree())
    ax.plot([-0.08, 132], [51.53, 43.17], transform=ccrs.Geodetic())

    plt.show()


if __name__ == '__main__':
    main()
```

```python
#Example from scitools.org eshowing built-in natural Earth features found in cartopy
def main():
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
    ax.set_extent([-20, 60, -40, 45], crs=ccrs.PlateCarree())

    ax.add_feature(cfeature.LAND) #shows land
    ax.add_feature(cfeature.OCEAN) #shows ocean
    ax.add_feature(cfeature.COASTLINE) #shows coastline
    ax.add_feature(cfeature.BORDERS, linestyle=':') #shows borders
    ax.add_feature(cfeature.LAKES, alpha=0.5) #shows lakes
    ax.add_feature(cfeature.RIVERS) #shows rivers

    plt.show()


if __name__ == '__main__':
    main()
```
